```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data.csv", encoding="latin1")
```

```python
x=df["no2"].dropna().values
```

```python
r=102303514
```

```python
ar=0.05*(r%7)
br=0.3*((r%5)+1)
```

```python
z=x+ar*np.sin(br*x)
```

```python
mu=np.mean(z)
```

```python
var=np.var(z)
```

```python
lambda_val = 1/ (2*var)
```

```python
c=1/np.sqrt(2*np.pi*var)
```

```python
pdf=c*np.exp(-lambda_val*(np.sort(z)-mu)**2)
```

```python
print("Estimated Paramters are:")
print("mu:", mu)
print("lambda:", lambda_val)
print("c:", c)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.hist(z, bins=40, density=True)
plt.plot(np.sort(z), pdf, color="red")
plt.show()
```
